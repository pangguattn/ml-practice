# Feature Selection

```python
import pandas as pd

# Run this cell if you
data = pd.read_csv("https://wagon-public-datasets.s3.amazonaws.com/05-Machine-Learning/02-Prepare-the-dataset/cleaned_dataset.csv")

data.head()
```

## Correlation investigation

```python
import seaborn as sns

correlation_matrix = data.corr()
column_names= correlation_matrix.columns
sns.heatmap(correlation_matrix, xticklabels=column_names,  yticklabels=column_names,cmap= "bwr", annot=True, fmt='.1g');
```

👉 None of pairs exceed a correlation of 0.9 or -0.9 

## Base Modelling

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate

y = data['SalePrice']

# Defining the features
X = data.drop(columns=['SalePrice'])

# Instantiate model
model = LinearRegression().fit(X, y)

# Scoring on multiple folds aka Cross Validation
cv_results = cross_validate(model, X, y, cv=10)
base_model_score = cv_results["test_score"].mean()
base_model_score
```

##  Feature Permutation

```python
from sklearn.inspection import permutation_importance
import numpy as np

# Fit model
model = LinearRegression().fit(X, y) 

# Performs Permutation
permutation_score = permutation_importance(model, X, y, n_repeats=10) 

# Unstack results showing the decrease in performance after shuffling features
importance_df = pd.DataFrame(np.vstack((X.columns,
                                        permutation_score.importances_mean)).T) 

importance_df.columns=['feature','score decrease']

# Show the important features
importance_df.sort_values(by="score decrease", ascending = False) 
```

👉 The most important feature is GrLivArea.

## Modelling with less complexity

👉 Dropping features one by one starting from the ones lowest importance until your model score starts dropping significantly.

```python
data_drop = data.drop(columns=['RoofSurface', 'OverallCond'])
data_drop


y = data_drop['SalePrice']

# Defining the features
X = data_drop.drop(columns=['SalePrice'])

# Instantiate model
model = LinearRegression().fit(X, y)

# Scoring on multiple folds aka Cross Validation
cv_results = cross_validate(model, X, y, cv=10)
simplified_model_score = cv_results["test_score"].mean()
simplified_model_score
```
